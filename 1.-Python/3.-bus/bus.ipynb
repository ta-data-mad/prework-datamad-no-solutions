# Bus

This bus has a passenger entry and exit control system to monitor the number of occupants it carries and thus detect when there is too high a capacity.

At each stop the entry and exit of passengers is represented by a tuple consisting of two integer numbers.
```
bus_stop = (in, out)
```
The succession of stops is represented by a list of these tuples.
```
stops = [(in1, out1), (in2, out2), (in3, out3), (in4, out4)]
```

## Goals:
* lists, tuples
* while/for loops
* minimum, maximum, length
* average, standard deviation

## Tasks
1. Calculate the number of stops.
2. Assign to a variable a list whose elements are the number of passengers at each stop (in-out),
3. Find the maximum occupation of the bus.
4. Calculate the average occupation. And the standard deviation.


````python
# variables

stops = [(10, 0), (4, 1), (3, 5), (3, 4), (5, 1), (1, 5), (5, 8), (4, 6), (2, 3)]
````

````python
# 1. Calculate the number of stops.
len(stops)
````

**Expected output**:
```
9
```

````python
# 2. Assign a variable a list whose elements are the number of passengers in each stop: 
# Each item depends on the previous item in the list + in - out.
pax=0
pax_each_stop=[]
for pax_in, pax_out in stops:
    
    pax += pax_in - pax_out
    
    pax_each_stop.append(pax)

print(pax_each_stop)
````

**Expected output**:
```
[10, 13, 11, 10, 14, 10, 7, 5, 4]
```

````python
# 3. Find the maximum occupation of the bus.
max(pax_each_stop)
````

**Expected output**:
```
14
```

````python
# 4. Calculate the average occupation. And the standard deviation.
avg = sum(pax_each_stop) / len(pax_each_stop)

print("The average occupation is", avg, "passengers.\n")


calc_sd = []
for i in pax_each_stop:
    calc_sd.append((i-avg)**2)
    
sd_occupation=((sum(calc_sd))/len(pax_each_stop))**(1/2)

print("The standard desviation is", sd_occupation,".")
````

**Expected output**:
```
9.333333333333334
3.197221015541813
```

````python
import statistics
print("The average occupation is", statistics.mean(pax_each_stop) , "passengers.\n")
print("The standard desviation is", statistics.pstdev(pax_each_stop),".")
````
