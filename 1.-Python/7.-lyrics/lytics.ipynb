### Step 1. Create bottles_lyrics function

````python
def bottles_lyrics():
    a = str("bottles of beer on the wall")
    b = str("bottles of beer.\n")
    c = str("Take one down and pass it around,")
    d = 100
    e = d-1
    f = str("2 bottles of beer on the wall, 2 bottles of beer.\nTake one down and pass it around, 1 bottle of beer on the wall.\n\n1 bottle of beer on the wall, 1 bottle of beer.\nTake one down and pass it around, no more bottles of beer on the wall.\n\nNo more bottles of beer on the wall, no more bottles of beer.\nGo to the store and buy some more, 99 bottles of beer on the wall.")
    result=""
    
    for i in range(2,99):
        d-=1
        e = d-1
        result+=("{3} {0}, {3} {1}{2} {4} {0}.\n\n".format(a,b,c,d,e))
    result+=(f)
    return result   
    # your code here. 
````
