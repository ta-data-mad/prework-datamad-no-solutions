### Step 4: Function
 Create a functions called babyshark() that generates the wanted output. 

````python
def baby_shark_lyrics():     
    song = []
    song_names = ["""Baby Shark """, """Mummy Shark """, """Daddy Shark """, "Grandma Shark """, """Grampa Shark """, """Lets'go hunt """, """Run Away,..."""]
    for baby in song_names[0:6]: 
        i = baby + """doo """*6
        j = i*3 + baby + "!"
        song.append(j)
    song.append(song_names[6] + "!")
    return(song)
    

    
  
baby_shark_lyrics()    
````
