# Duel of sorcerers
## Intro
You are witnessing an epic battle between two powerful sorcerers: Gandalf and Saruman. Each sorcerer has 10 spells of variable power in their mind and they are going to throw them one after the other. The winner of the duel will be the one who wins more of those clashes between spells. Spells are represented as a list of 10 integers whose value equals the power of the spell.
```
gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]
```
For example:
1. The first clash is won by Saruman: 10 against 23, wins 23
2. The second clash wins Saruman: 11 against 66, wins 66
3. etc.


You will create two variables, one for each sorcerer, where the sum of clashes won will be stored. Depending on which variable is greater at the end of the duel, you will show one of the following three results on the screen:
* Gandalf wins
* Saruman wins
* Tie

## Solution

````python
# Assign spell power lists to variables
list_gandalf = [10, 11, 13, 30, 22, 11, 10, 33, 22, 22]
list_saruman = [23, 66, 12, 43, 12, 10, 44, 23, 12, 17]
len(list_gandalf)
len(list_saruman)
````

````python
# Assign 0 to each variable that stores the victories

clashes_gandalf=0
clashes_saruman=0
````

````python
# Execution of spell clashes

result=zip(list_gandalf,list_saruman)
result_list=list(result)
print(result_list)

for list_gandalf,list_saruman in result_list:
    if list_gandalf>list_saruman:
        clashes_gandalf+=1
    elif list_gandalf<list_saruman:
        clashes_saruman+=1
print("Total gandalf wins:", clashes_gandalf)
print("Total saruman wins:", clashes_saruman)
````

**Expected output**:
```
6
4
```

````python
# We check who has won, do not forget the possibility of a draw.
# Print the result based on the winner.
if clashes_gandalf>clashes_saruman:
    print("Gandalf wins")
elif clashes_gandalf<clashes_saruman:
    print("Saruman wins")
````

**Expected output**:
```
Gandalf wins
```

````python
# 1. Spells now have a name and there is a dictionary that relates that name to a power.
# variables

POWER = {
    'Fireball': 50, 
    'Lightning bolt': 40, 
    'Magic arrow': 10, 
    'Black Tentacles': 25, 
    'Contagion': 45
}

gandalf = ['Fireball', 'Lightning bolt', 'Lightning bolt', 'Magic arrow', 'Fireball', 
           'Magic arrow', 'Lightning bolt', 'Fireball', 'Magic arrow', 'Fireball']
saruman = ['Contagion', 'Contagion', 'Black Tentacles', 'Fireball', 'Black Tentacles', 
           'Lightning bolt', 'Magic arrow', 'Contagion', 'Magic arrow', 'Magic arrow']
````

````python
# Assign spell power lists to variables

gandalf_power=[]
saruman_power=[]

for power in gandalf:
    gandalf_power.append(POWER[power])
    
for power in saruman:
    saruman_power.append(POWER[power])
    
print(gandalf_power)
print(saruman_power)

clashes=zip(gandalf_power,saruman_power)
clashes_list=list(clashes)
print(clashes_list)
````

**Expected output**:
```
([50, 40, 40, 10, 50, 10, 40, 50, 10, 50],
 [45, 45, 25, 50, 25, 40, 10, 45, 10, 10])
 ```

````python
# 2. A sorcerer wins if he succeeds in winning 3 spell clashes in a row.



# Execution of spell clashes

# check clashes

# check for 3 wins in a row

# check the winner

# count sum of clashes

gandalf_wins=0
saruman_wins=0

for gandalf_power,saruman_power in clashes_list:
    if gandalf_power>saruman_power:
        gandalf_wins+=1
        saruman_wins+=0
    elif gandalf_power<saruman_power:
        gandalf_wins+=0
        saruman_wins+=1
    elif gandalf_power==saruman_power:
        gandalf_wins+=0
        saruman_wins+=0
        
print("Gandalf:", gandalf_wins)
print("Saruman:", saruman_wins)

if gandalf_wins>saruman_wins:
    print("Gandalf wins")
elif gandalf_wins<saruman_wins:
    print("Saruman wins")
````

**Expected output**:
```
Gandalf wins
```

````python
# 3. Average of each of the spell lists.

avg_gandalf=sum(gandalf_power)/len(gandalf_power)
avg_saruman=sum(saruman_power)/len(saruman_power)

print("Gandalf average:", avg_gandalf)
print("Saruman average:", avg_saruman)
````
